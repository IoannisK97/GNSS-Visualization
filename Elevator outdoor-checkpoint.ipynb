```python
def turnLineToNum(lineList):
    newLine=[]
    for i in range(1,len(lineList)):
        if lineList[i]!='' and lineList[i]!=' ':
            newLine.append(float(lineList[i]))
        else:
            newLine.append(float(0))
    newLine=[lineList[0]]+newLine
    return newLine

def medianList(myList):
    median=statistics.median(myList)
    return [median]*len(myList)

def fileTraversing(fileName):
    f=open(fileName)
    listOfFile=[]
    for line in f:
        line=line.strip()
        if len(line)>0:
            if line[0]!="#":
                splitLine=line.split(",")

                listOfFile.append(splitLine)
    
    listOfStatus=[]

    listOfRaw=[]

    listOfMagn=[]

    for i in range(0,len(listOfFile)):
        
        record=listOfFile[i]

        if record[0]=="Status":
            listOfStatus.append(turnLineToNum(record))
        elif record[0]=="Raw":
            listOfRaw.append(turnLineToNum(record))

        elif record[0]=="UncalMag":
            listOfMagn.append(turnLineToNum(record))

    SNR=[]
    accumulatedState=[]
    for record in listOfRaw:

        SNR.append(record[27])
        accumulatedState.append(record[19])
    
    noConstellations=[]
    satelliteID=[]

    for record in listOfStatus:
        noConstellations.append(record[2])
        satelliteID.append(record[5])

    
    
        
    magnX=[]
    magnY=[]
    magnZ=[]
    for record in listOfMagn:
        magnX.append(record[3])
        magnY.append(record[4])
        magnZ.append(record[5])

    
    
    returnDict={"MagnetometerZ":magnZ,"MagnetometerY":magnY,"MagnetometerX":magnX,"ConstellationsNum":noConstellations,"SignalToNoise":SNR, 'satelliteID':satelliteID,'accumState':accumulatedState}
    
    return returnDict

xiaomiIndoorNearElevator=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\xiaomi indoor near elevator.txt")
xiaomiIndoorFarElevator=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\xiaomi indoor far elevator.txt")
xiaomiOutdoorFar=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\xiaomi outdoor far.txt")
samsungIndoorNearElevator=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\samsung indoor near elevator.txt")
samsungIndoorFarElevator=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\samsung indoor far elevator.txt")
samsungOutdoorFar=fileTraversing("C:\\Users\\user\\OneDrive\\Υπολογιστής\\IndoorOutdoor\\28June measurementd\\samsung outdoor far.txt")

```

```python
import matplotlib.pyplot as plt
import statistics

def createXaxis(myList):
    return range(0,len(myList))
def plotAGraph(measureList,yLabel,Title,color,scatter=False,xLabel='Time',si=1):
    if (scatter==False):
        x=measureList
        plt.ylabel(yLabel)
        plt.xlabel(xLabel)
        plt.title(Title)
        
        plt.plot(x,c=color)
        plt.show()
        
    else:
        plt.scatter(createXaxis(measureList),measureList,c=color,s=si)
        plt.ylabel(yLabel)
        plt.xlabel(xLabel)
        plt.title(Title)  
        plt.show()
        
def plotMedians(nearList,farList,outList,yLabel,Title,xLabel="Time"):
    plt.plot(medianList(nearList),c='r')
    plt.plot(medianList(farList),c='g')
    plt.plot(medianList(outList),c='b')
    plt.ylabel(yLabel)
    plt.title(Title)
```
